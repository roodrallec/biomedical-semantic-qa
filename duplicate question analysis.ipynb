```python
import pandas as pd
import pickle
import part_1
from keras.preprocessing.sequence import pad_sequences
```

```python
df = pd.DataFrame.from_csv('./datasets/quora_duplicate_questions.tsv', sep='\t')
```

```python
duplicates = df[df['is_duplicate'] == 1]
```

```python
MODEL_DIR="./Graph/2019-01-04 10:43:55.028986/model.pickle"
```

```python
with open(MODEL_DIR, 'rb') as f:
    model = pickle.load(f)
```

```python
model.summary()
```

```python
part_1_vocab = part_1.get_vocab_idx()
part_1_seq_length = model.layers[0].get_output_at(0).get_shape().as_list()[1] 
part_2_sequences = []

for text in duplicates['question1']:
    sequence = []    
    missing_token = False
    tokens = part_1.text_to_tokens(text)
    
    for token in tokens:
        if token not in part_1_vocab.keys():
            missing_token = True
        else:
            sequence.append(part_1_vocab[token])

    if missing_token or len(sequence) > part_1_seq_length:
        continue
    else:
        part_2_sequences.append(sequence)    
```
